# Logistic Regression

```python
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())
```

```python
import numpy as np
X = iris["data"][:, 3:] # petal width
y = (iris["target"] == 2).astype(np.int64) 
```

```python
from sklearn.linear_model import LogisticRegression
logisitic  = LogisticRegression().fit(X,y)
```

```python
X_new = np.linspace(0,3,num=1000).reshape(-1, 1)
```

```python
pred_proba = logisitic.predict_proba(X_new)
```

```python
import matplotlib.pyplot as plt
plt.plot(X_new, pred_proba[:, 1], "g-", label="Iris virginica")
plt.plot(X_new, pred_proba[:, 0], "b--", label="Not Iris virginica")
plt.legend()
```

Therefore, there is a decision
boundary at around 1.6 cm where both probabilities are equal to 50%: if the petal
width is higher than 1.6 cm, the classifier will predict that the flower is an Iris virgin‚Äê
ica, and otherwise it will predict that it is not (even if it is not very confident)
