```python
from sklearn.mixture import BayesianGaussianMixture
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=1000, noise=0.05)
bgm = BayesianGaussianMixture(n_components=10, random_state=42)
bgm.fit(X)
bgm.weights_
```
