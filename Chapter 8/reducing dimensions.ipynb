# Principle components

### Principle:
Principle components are the axis on to which the data can be projected as in real weorld case most of the feature are constant or corelated and do not introduce new direction and thus can be projected on to a reduced dimension  

```python
import numpy as np
```

```python
X = np.array([
    [2, 4],
    [4, 2],
    [6, 8],
    [8, 6]
])
```

```python
X_centered= X - X.mean(axis = 0)
X_centered
```

```python
U,s,Vt = np.linalg.svd(X_centered)
```

```python
c1 = Vt.T[:, 0]
c2 = Vt.T[:, 1]
```

```python
c1,c2
```

### Reducing to d dimensions


```python
Wd = Vt[:,:2]
Xproj = X.dot(Wd)
Xproj
```

### Using scikit learn

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
Xproj =  pca.fit_transform(X)
```

```python
Xproj
```

After fitting the PCA transformer to the dataset, its components_ attribute holds the
transpose of Wd (e.g., the unit vector that defines the first principal component is
equal to pca.components_.T[:, 0]

```python
pca.components_
```

```python
Wd.T
```

Another useful piece of information is the explained variance ratio of each principal
component, available via the explained_variance_ratio_ variable

```python
pca.explained_variance_ratio_
```
