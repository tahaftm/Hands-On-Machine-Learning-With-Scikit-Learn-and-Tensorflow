```python
import numpy as np
def onehot(y):
    zeros = np.zeros((len(y), len(np.unique(y))))
    zeros[np.arange(len(y)), y] = 1
    return zeros

def train(X,y,n_iterations=1000,eta=0.01, epoch = 1000):
    m,n=X.shape
    X_b =np.c_[np.ones((m,1)),X]
    theta = np.random.randn(X_b.shape[1],len(np.unique(y)))*0.01
    y_one_hot = onehot(y)
    for i in range(n_iterations):
        pred = X_b@theta
        score = softmax_scores(pred)
        error = score - y_one_hot
        loss = -np.mean(np.sum(y_one_hot * np.log(score + 1e-9), axis=1))
        
        grad = (1/m) * X_b.T @ (error)
        theta = theta - eta * grad
        
        if epoch % 100 == 0:
            print("Loss:", loss)
            best_loss = float('inf')
        counter = 0

        if loss < best_loss:
            best_loss = loss
            counter = 0
        else:
            counter += 1

        if counter == 10:
            break
    return theta

def softmax_scores(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))  
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def softmax(X,theta):
    m,n = X.shape
    X_b = np.c_[np.ones((m,1)),X]
    score = X_b@theta
    return softmax_scores(score)

def predict(X, theta):
    probs = softmax(X, theta)
    return np.argmax(probs, axis=1)
```

```python
def generate_dataset(n_samples=300):
    np.random.seed(42)

    X = []
    y = []

    # Class 0
    X0 = np.random.randn(n_samples, 2) + np.array([0, 0])
    y0 = np.zeros(n_samples)

    # Class 1
    X1 = np.random.randn(n_samples, 2) + np.array([5, 5])
    y1 = np.ones(n_samples)

    # Class 2
    X2 = np.random.randn(n_samples, 2) + np.array([0, 5])
    y2 = np.full(n_samples, 2)

    X = np.vstack([X0, X1, X2])
    y = np.concatenate([y0, y1, y2])

    return X, y.astype(int)

X, y = generate_dataset()
```
