```python
import numpy as np

X = np.array([
    [1, 2],
    [2, 1],
    [1.5, 1.8],

    [5, 6],
    [6, 5],
    [5.5, 5.2],

    [9, 1],
    [8, 2],
    [9.5, 1.5]
])

y = np.array([
    0, 0, 0,   # Class 0
    1, 1, 1,   # Class 1
    2, 2, 2    # Class 2
])
```

```python
def one_hot(y):
    encoded = np.zeros((y.shape[0], len(np.unique(y))))
    encoded[np.arange(len(y)), y] = 1
    return encoded

def train_softmax(X,y,n_iterations = 5000, eta  = 0.01):
    m,n = X.shape
    X_b = np.c_[np.ones((m,1)),X]
    num_classes = len(np.unique(y))
    theta = np.zeros((n+1, num_classes))   
    y_one_hot = one_hot(y)
    for i in range(n_iterations): 
        z = X_b.dot(theta)
        predictions = softmax_scores(z)
        error = predictions - y_one_hot
        grad = (1/m) * X_b.T.dot(error)
        theta = theta - eta*grad
    return theta

def softmax_scores(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))  
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def softmax(X,theta):
    m,n = X.shape
    X_b = np.c_[np.ones((m,1)),X]
    score = X_b@theta
    return softmax_scores(score)

def predict(X, theta):
    probs = softmax(X, theta)
    return np.argmax(probs, axis=1)
```
