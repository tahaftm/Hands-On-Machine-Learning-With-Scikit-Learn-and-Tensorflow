```python
from sklearn.datasets import load_iris
data = load_iris()
data
```

```python
import pandas as pd
X =data["data"][:, (2, 3)]
y = data['target']
```

```python
X
```

```python
y
```

```python
import numpy as np
np.unique_counts(y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], marker='x', c='red')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC, SVC
linear_svc = LinearSVC(loss="hinge", C = 1)
linear_params = {'C': [0.1,1,10,100,150]}
linear_grid = GridSearchCV(linear_svc, param_grid=linear_params, cv = 5, scoring='accuracy')
linear_grid.fit(X_test,y_test)
```

```python
linear_grid.best_score_
```

```python
svc = SVC(kernel='linear', C=1)
svc_grid = GridSearchCV(svc, param_grid=linear_params, cv = 5, scoring='accuracy')
svc_grid.fit(X_train,y_train)
```

```python
svc_grid.best_score_
```

```python
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier(loss='hinge').fit(X_train,y_train)
sgd_pred = sgd.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, sgd_pred)
```
