```python
import numpy as np
def train_equation(X,y, eta = 0.01, n_iterations = 5000):
    m,n = X.shape
    y = y.reshape(-1,1)
    X_b = np.c_[np.ones((m,1)), X]
    theta = np.ones((n+1,1))
    for i in range(n_iterations):
        z = X_b.dot(theta)
        predictions = logisitic(z)
        error = predictions - y
        grad = (1/m) * X_b.T.dot(error)
        theta = theta - eta*grad
    return theta

def predict(X,theta):
    m,n = X.shape
    X_b = np.c_[np.ones((m,1)), X]
    return logisitic(X_b.dot(theta))

def logisitic(t):
    return 1/(1+np.exp(-t))
```

```python
X = np.array([
    [0.5, 1.5],
    [1.0, 2.0],
    [1.5, 0.5],
    [2.0, 1.0],
    [3.0, 3.5],
    [3.5, 4.0],
    [4.0, 3.0],
    [4.5, 4.5]
])

# Labels: y (0 or 1)
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])
probablity = logisitic(predict(X, train_equation(X,y)))
probablity
```

```python
decision_boundary = 0.5
classes = []
for i in probablity:
    if i < decision_boundary:
        classes.append(0)
    else:
        classes.append(1)
classes
```

### Training a cost function

```python
def log_loss(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)

    loss = -np.mean(
        y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)
    )    
    return loss
```

```python
log_loss(y, probablity)
```
