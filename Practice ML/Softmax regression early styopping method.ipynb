Implement Batch Gradient Descent with early stopping for Softmax Regression
(without using Scikit-Learn).

Generate datasets

```python
import numpy as np

def generate_dataset(n_samples=300):
    np.random.seed(42)

    X = []
    y = []

    # Class 0
    X0 = np.random.randn(n_samples, 2) + np.array([0, 0])
    y0 = np.zeros(n_samples)

    # Class 1
    X1 = np.random.randn(n_samples, 2) + np.array([5, 5])
    y1 = np.ones(n_samples)

    # Class 2
    X2 = np.random.randn(n_samples, 2) + np.array([0, 5])
    y2 = np.full(n_samples, 2)

    X = np.vstack([X0, X1, X2])
    y = np.concatenate([y0, y1, y2])

    return X, y.astype(int)

X, y = generate_dataset()
```

```python
def softmax_score(x, theta_k):
    return np.dot(x,theta_k)
```

```python
def softmax_function(scores):
    scores = scores - np.max(scores, axis=1, keepdims=True)
    exp_scores = np.exp(scores)
    return exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
```

```python
def one_hot(y,  num_classes):
    encoded  = np.zeros((len(y), num_classes))
    encoded[np.arange(len(y)), y] = 1
    return encoded
```

```python
def add_bias(X):
    return np.c_[np.ones((X.shape[0], 1)), X]
```

```python
def train(X, y, lr=0.1, epochs=1000):
    X = add_bias(X)
    m, n = X.shape
    k = len(np.unique(y))
    
    Y = one_hot(y, k)
    Theta = np.random.randn(n, k) * 0.01
    
    for epoch in range(epochs):
        scores = X @ Theta
        probs = softmax_function(scores)
        
        loss = -np.mean(np.sum(Y * np.log(probs + 1e-9), axis=1))
        
        grad = (1/m) * X.T @ (probs - Y)
        Theta = Theta - lr * grad
        
        if epoch % 100 == 0:
            print("Loss:", loss)
            best_loss = float('inf')
        counter = 0

        if loss < best_loss:
            best_loss = loss
            counter = 0
        else:
            counter += 1

        if counter == 10:
            break
    
    return Theta
```

```python
def predict(X, Theta):
    X = add_bias(X)
    probs = softmax_function(X @ Theta)
    return np.argmax(probs, axis=1)
```

```python
X,y = generate_dataset()

Theta = train(X, y)

preds = predict(X, Theta)

accuracy = np.mean(preds == y)

print("\nFinal Accuracy:", accuracy)
```
