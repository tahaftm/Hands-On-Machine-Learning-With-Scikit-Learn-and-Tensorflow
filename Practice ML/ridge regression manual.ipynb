### Closed form equation



```python
import numpy as np
def train_equation(X, y):
    alpha = 0.1
    m = X.shape[0]
    X_b = np.c_[np.ones((m,1)), X]
    n = X_b.shape[1]
    A = np.identity(n)
    A[0][0] = 0
    theta = (np.linalg.inv(X_b.T.dot(X_b) + (alpha*A))).dot((X_b.T.dot(y)))
    return theta

def predict(X, theta):
    m,n = X.shape
    X_b = np.c_[np.ones((m,1)), X]
    return X_b.dot(theta) 
```

```python
import numpy as np
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
X.shape
```

```python
theta = train_equation(X,y)
predict(X,theta)
```

### Gradient Descent

```python
def train_grad_descent(X,y):
    m,n = X.shape
    eta = 0.01
    alpha = 0.1
    n_iterations = 1000
    X_b = np.c_[np.ones((m,1)), X]
    theta = np.random.randn(X_b.shape[1],1)
    for i in range(n_iterations):
        y_pred = X_b @ theta
        gradient = (2/m)* (X_b.T @ (y_pred - y))
        reg = 2*alpha*theta
        reg[0] = 0
        gradient += reg
        theta = theta - eta * gradient
    return theta
```

```python
theta = train_grad_descent(X,y)
theta
```

```python
predict(X,theta)
```

### SGD

```python
t0,t1 = 5, 50
def learning_schedule(t):
    return t0 / (t1+t)

def train_SGD(X,y,n_epochs = 50, n_iterations = 1000,alpha = 0.1):
    m,n = X.shape
    X_b = np.c_[np.ones((m,1)),X]
    theta = np.random.randn(X_b.shape[1], 1)
    for epoch in range(n_epochs):
        for i in range(n_iterations):
            random_index = np.random.randint(m)
            random_instance = X_b[random_index:random_index+1]
            random_pred = y[random_index:random_index+1]
            gradients = 2 * random_instance.T.dot(random_instance.dot(theta) - random_pred)
            reg = 2 * alpha * theta
            reg[0] = 0
            gradients += reg
            eta = learning_schedule(epoch * m + i)
            theta = theta - gradients * eta
    return theta
```

```python
theta = train_SGD(X,y)
```

```python
predict(X,theta)
```
