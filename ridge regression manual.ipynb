### Closed form equation



```python
import numpy as np
def train_equation(X, y):
    alpha = 0.1
    m = X.shape[0]
    X_b = np.c_[np.ones((m,1)), X]
    n = X_b.shape[1]
    A = np.identity(n)
    A[0][0] = 0
    theta = (np.linalg.inv(X_b.T.dot(X_b) + (alpha*A))).dot((X_b.T.dot(y)))
    return theta

def predict(X, theta):
    m,n = X.shape
    X_b = np.c_[np.ones((m,1)), X]
    return X_b.dot(theta) 
```

```python
import numpy as np
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
X.shape
```

```python
theta = train_equation(X,y)
predict(X,theta)
```

### Gradient Descent

```python
def train_grad_descent(X,y):
    m,n = X.shape
    eta = 0.01
    alpha = 0.1
    n_iterations = 1000
    X_b = np.c_[np.ones((m,1)), X]
    theta = np.random.randn(X_b.shape[1],1)
    for i in range(n_iterations):
        y_pred = X_b @ theta
        gradient = (2/m)* (X_b.T @ (y_pred - y))
        reg = 2*alpha*theta
        reg[0] = 0
        gradient += reg
        theta = theta - eta * gradient
    return theta
```

```python
theta = train_grad_descent(X,y)
theta
```

```python
predict(X,theta)
```
