# Linear Regression


### Downloading the training and testing data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("C:\\Users\\DELL\\Desktop\\Hands on machine learning\\Chapter 5\\Exercises\\regression\\data\\training data.csv")
train_data.info()
```

```python
test_data = pd.read_csv("C:\\Users\\DELL\\Desktop\\Hands on machine learning\\Chapter 5\\Exercises\\regression\\data\\testing data.csv")
test_data.info()
```

##### Dividing into dependent and independent variables

```python
train_data.head()
```

```python
X_train = train_data.drop(labels = ['median_house_value'],axis = 1)
X_train.head()
```

```python
X_test = test_data.drop(labels = ['median_house_value'],axis = 1)
X_test.head()
```

```python
y_train = train_data['median_house_value']
y_train
```

```python
y_test = test_data['median_house_value']
y_test
```

### Training Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("Mean Absolute Error: ", mean_absolute_error(y_test,y_pred))
print("Mean Squared Error: ", mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error: ", np.sqrt(mean_absolute_error(y_test,y_pred)))
```

```python
from sklearn.metrics import r2_score
print("R square: ", r2_score(y_test,y_pred))

n = X_test.shape[0]  
p = X_test.shape[1]  

adj_r2 = 1 - (1 - r2_score(y_test,y_pred)) * (n - 1) / (n - p - 1)

print("Adjusted R square: ", adj_r2)
```

### Hyperparameter Tuning

```python
params = {
    'fit_intercept' : [True,False],     ## Keep it false only when the data is centered. Which mean that average valuye of each feature must be 0 this can be done by (X_centered = X - X.mean())
    'copy_X' : [True, False],           ## In some technique linear regression might use such as graident descent the model sometime need top changer the original traing data
    'positive' : [True, False]          ## To neccassarily have al positive coefficients
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(model, param_grid=params, cv = 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
new_model = grid.best_estimator_
```

```python
pred_new = new_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("Mean Absolute Error: ", mean_absolute_error(y_test,pred_new))
print("Mean Squared Error: ", mean_squared_error(y_test,pred_new))
print("Root Mean Squared Error: ", np.sqrt(mean_absolute_error(y_test,pred_new)))
```

```python
from sklearn.metrics import r2_score
print("R square: ", r2_score(y_test,pred_new))

n = X_test.shape[0]  
p = X_test.shape[1]  

adj_r2 = 1 - (1 - r2_score(y_test,pred_new)) * (n - 1) / (n - p - 1)

print("Adjusted R square: ", adj_r2)
```
