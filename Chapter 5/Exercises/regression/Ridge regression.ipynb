# Ridge Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("C:\\Users\\DELL\\Desktop\\Hands on machine learning\\Chapter 5\\Exercises\\regression\\data\\training data.csv")
train_data.info()
```

```python
test_data = pd.read_csv("C:\\Users\\DELL\\Desktop\\Hands on machine learning\\Chapter 5\\Exercises\\regression\\data\\testing data.csv")
test_data.info()
```

```python
X_train = train_data.drop(labels = ['median_house_value'],axis = 1)
X_train.head()
```

```python
X_test = test_data.drop(labels = ['median_house_value'],axis = 1)
X_test.head()
```

```python
y_train = train_data['median_house_value']
y_train
```

```python
y_test = test_data['median_house_value']
y_test
```

### Training the model

```python
from sklearn.linear_model import Ridge
model = Ridge(solver='cholesky')
```

```python
model.fit(X_train,y_train)
```

```python
ridge_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("Mean Absolute Error: ", mean_absolute_error(y_test,ridge_pred))
print("Mean Squared Error: ", mean_squared_error(y_test,ridge_pred))
print("Root Mean Squared Error: ", np.sqrt(mean_absolute_error(y_test,ridge_pred)))
```

```python
from sklearn.metrics import r2_score
print("R square: ", r2_score(y_test,ridge_pred))

n = X_test.shape[0]  
p = X_test.shape[1]  

adj_r2 = 1 - (1 - r2_score(y_test,ridge_pred)) * (n - 1) / (n - p - 1)

print("Adjusted R square: ", adj_r2)
```

### Hyperparamter tuning

```python
param = {
    'alpha' : [1,2,3,4,5,6,7,8,9]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(model, param_grid=param, cv =3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("Mean Absolute Error: ", mean_absolute_error(y_test,grid_pred))
print("Mean Squared Error: ", mean_squared_error(y_test,grid_pred))
print("Root Mean Squared Error: ", np.sqrt(mean_absolute_error(y_test,grid_pred)))
```

```python
from sklearn.metrics import r2_score
print("R square: ", r2_score(y_test,grid_pred))

n = X_test.shape[0]  
p = X_test.shape[1]  

adj_r2 = 1 - (1 - r2_score(y_test,grid_pred)) * (n - 1) / (n - p - 1)

print("Adjusted R square: ", adj_r2)
```
