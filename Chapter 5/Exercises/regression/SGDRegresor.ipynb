# SGD Regressor

### Downloading the training and testing data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("C:\\Users\\DELL\\Desktop\\Hands on machine learning\\Chapter 5\\Exercises\\regression\\data\\training data.csv")
train_data.info()
```

```python
test_data = pd.read_csv("C:\\Users\\DELL\\Desktop\\Hands on machine learning\\Chapter 5\\Exercises\\regression\\data\\testing data.csv")
test_data.info()
```

##### Dividing into dependent and independent variables

```python
train_data.head()
```

```python
X_train = train_data.drop(labels = ['median_house_value'],axis = 1)
X_train.head()
```

```python
X_test = test_data.drop(labels = ['median_house_value'],axis = 1)
X_test.head()
```

```python
y_train = train_data['median_house_value']
y_train
```

```python
y_test = test_data['median_house_value']
y_test
```

### Training Model

```python
from sklearn.linear_model import SGDRegressor
model = SGDRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("Mean Absolute Error: ", mean_absolute_error(y_test,y_pred))
print("Mean Squared Error: ", mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error: ", np.sqrt(mean_absolute_error(y_test,y_pred)))
```

```python
from sklearn.metrics import r2_score
print("R square: ", r2_score(y_test,y_pred))

n = X_test.shape[0]  
p = X_test.shape[1]  

adj_r2 = 1 - (1 - r2_score(y_test,y_pred)) * (n - 1) / (n - p - 1)

print("Adjusted R square: ", adj_r2)
```

### Hyperparameter Tuning

```python
params = {
    'loss': ['squared_error', 'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive'],
    'penalty': ['l2', 'l1', 'elasticnet', None],
    'alpha': [0.0001, 0.001, 0.01, 0.1],        # regularization strength
    'l1_ratio': [0, 0.15, 0.5, 0.85, 1],       # only used if penalty='elasticnet'
    'fit_intercept': [True, False],
    'max_iter': [1000, 2000, 5000],
    'tol': [1e-3, 1e-4, 1e-5],
    'shuffle': [True, False],
    'epsilon': [0.1, 0.2, 0.5],                # only used for huber or epsilon-insensitive
    'learning_rate': ['constant', 'optimal', 'invscaling', 'adaptive'],
    'eta0': [0.01, 0.1, 0.5],                  # initial learning rate
    'power_t': [0.25, 0.5, 0.75],              # only used for 'invscaling'
    'early_stopping': [True, False],
    'validation_fraction': [0.1, 0.2],         # only if early_stopping=True
    'n_iter_no_change': [5, 10]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(model, param_grid=params, cv = 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```
